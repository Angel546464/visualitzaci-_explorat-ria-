# VISUALIZACIONES

  ### Variable categórica Distrito

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 10))
ax = sns.countplot(y='Nom_Districte', data=preu_lloguer_df, order=preu_lloguer_df['Nom_Districte'].value_counts().index)
plt.title('Cantidad de Entradas por Barrio')
plt.xlabel('Cantidad de Entradas')

for p in ax.patches:
    ax.annotate(f'{p.get_width()}', (p.get_width(), p.get_y() + p.get_height() / 2), ha='left', va='center', xytext=(5, 0), textcoords='offset points')

plt.show()
```

Conteo de las entradas por distrito. Horta-Guinardó, Nou Barris y Sant Marti son los que cuentan con mas datos estando los tres en 80. El que menos tiene es Les Corts con 24 y los cinco restantes estan entre 32 y 56.

### Variable numérica Precio (mensualidad)

```python
#para esta visualización uso el df con el precio por mes.

import numpy as np

percentiles_to_show = [0, 25, 50, 75, 100]
plt.figure(figsize=(10, 6))
sns.boxplot(x=preu_mensualidad_df['Preu'], color='green')
percentile_values = np.percentile(preu_mensualidad_df['Preu'], percentiles_to_show)
mean_value = np.mean(preu_mensualidad_df['Preu'])

for value, percentile in zip(percentile_values, percentiles_to_show):
    plt.axvline(x=value, linestyle='--', color='red', label=f'{percentile}th Percentile: {value:.2f}')

plt.axvline(x=mean_value, linestyle='-', color='blue', label=f'Mean: {mean_value:.2f}')
plt.legend()
plt.title('Boxplot de Precios de Alquiler por mensualidad')
plt.xlabel('Precio del Alquiler')
plt.show()
```

Este gráfico nos sirve para tener una vision general de los precios por mensualidad. La media de los precios se situa en 871.59€ mientras que la mediana se encuentra por debajo con un precio de 816.35€. El precio más bajo en el dataframe es de 447.50€ y el más alto 1816.50€. Entre el percentil 25 y el 75 hay menos variablidad en el precio abarcando una horquilla entre los 742.48€ al mes hasta los 934.52€ al mes.

### Variable numérica Precio (por superficie)

```python
# en este caso uso preu_superficie_df

percentiles_to_show = [0, 25, 50, 75, 100]
plt.figure(figsize=(10, 6))
sns.boxplot(x=preu_superficie_df['Preu'], color='green')
percentile_values = np.percentile(preu_superficie_df['Preu'], percentiles_to_show)
mean_value = np.mean(preu_superficie_df['Preu'])

for value, percentile in zip(percentile_values, percentiles_to_show):
    plt.axvline(x=value, linestyle='--', color='red', label=f'{percentile}th Percentile: {value:.2f}')

plt.axvline(x=mean_value, linestyle='-', color='blue', label=f'Mean: {mean_value:.2f}')
plt.legend()
plt.title('Boxplot de Precios de Alquiler por mensualidad')
plt.xlabel('Precio de Alquiler')
plt.show()
```

En este obtenemos un resultado parecido al anterior pero en este caso mostrando el precio del alquiler por metro cuadrado. El 50% de los datos (25% a 75%) están entre 11.62€/m2 y 13.60€/m2. # PROPUESTA DE VISULIZACIÓN
Conseguir una estimación de la media de metros cuadrados por piso en cada distrito. Comparar información con la media de los precios para comprobar donde se encuentran los pisos más grandes y los mas pequeños. 

Para ello calculo en los df por separado la media de los precios y mediante la funcion 'merge' de pandas los uno para poder dividir uno entre el otro y obtener la estimación.

```python
media_mensual_por_distrito = preu_mensualidad_df.groupby('Nom_Districte')['Preu'].mean().reset_index()
media_mensual_por_distrito.columns = ['Nom_Districte', 'Media_Precio_Mensual']

media_superficie_por_distrito = preu_superficie_df.groupby('Nom_Districte')['Preu'].mean().reset_index()
media_superficie_por_distrito.columns = ['Nom_Districte', 'Media_Precio_Superficie']

merged_df = pd.merge(media_mensual_por_distrito, media_superficie_por_distrito, on='Nom_Districte') 
merged_df['Estimacion_Superficie'] = merged_df['Media_Precio_Mensual'] / merged_df['Media_Precio_Superficie']
merged_df = merged_df.sort_values(by='Estimacion_Superficie', ascending=False)

merged_df
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Media_Precio_Mensual', y='Estimacion_Superficie', hue='Nom_Districte', data=merged_df, s=100)
plt.xlabel('Media Precio Mensual (Euros)')
plt.ylabel('Estimación de Superficie (m²)')
plt.title('Relación entre Media Precio Mensual y Estimación de Superficie por Distrito')
plt.legend(title='Distrito', fontsize='large')
plt.show()
```

Tanto al compronar el resultado de 'merged_df' como de este ultimo diagrama de dispersión que, efectivamente, la relación confirma que en general los pisos más grandes son los más caros y al revés.

```python
plt.switch_backend('agg')
plt.rcParams['figure.autolayout'] = True

import os
from PIL import Image

def exportar_graficos_en_notebook_como_png(directorio_salida='/Users/angel/Desktop/graficos'):
    os.makedirs(directorio_salida, exist_ok=True)

    plt.close('all')
    figuras = plt.get_fignums()

    for i, num_grafico in enumerate(figuras):
        plt.figure(num_grafico)

        nombre_archivo = f'grafico_{i + 1}.png'
        buf = plt.gcf().canvas.buffer_rgba()
        ancho, alto = plt.gcf().canvas.get_width_height()
        imagen = Image.frombytes("RGBA", (ancho, alto), buf)
        ruta_png = os.path.join(directorio_salida, nombre_archivo)

        imagen.save(ruta_png)
        print(f'Gráfico {num_grafico} exportado como {ruta_png}')

exportar_graficos_en_notebook_como_png(directorio_salida='/Users/angel/Desktop/graficos')
```
